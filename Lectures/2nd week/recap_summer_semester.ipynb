```python
import csv
import matplotlib.pyplot as plt

i = 0
x = []
y = []

with open('./data/Lee_data.csv', 'r') as file:
	data = csv.reader(file)
	next(data)
	data = list(data)

for row in data:
	x.append(row[0])
	y.append(row[2])

print(x)
print(y)

plt.figure(figsize=(5, 5), dpi=150)
plt.title('Car usage', size=10)
plt.scatter(x, y, color='r')
plt.legend(fontsize=10, loc='lower right')
plt.show()

```
