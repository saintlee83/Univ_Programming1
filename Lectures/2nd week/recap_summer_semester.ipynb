```python
import csv
import matplotlib.pyplot as plt

i = 0
x = []
y = []

with open('./data/Lee_data.csv', 'r') as file:
	data = csv.reader(file)
	next(data)
	data = list(data)

for row in data:
	x.append(row[0])
	y.append(row[2])

print(x)
print(y)

plt.figure(figsize=(5, 5), dpi=150)
plt.title('Car usage', size=10)
plt.scatter(x, y, color='r')
plt.legend(fontsize=10, loc='lower right')
plt.show()

```

```python
import csv

with open('./data/card.csv', 'r', encoding='utf-8') as file:
	data = csv.reader(file)
	next(data)
	data = list(data)

print(data)

import matplotlib.pyplot as plt

s_mon = [0,0,0]

for row in data :
	if row[-1] == '전표매입' :
		mon, payment = int(row[0].split('/')[1]), int(row[-3])
		idx = mon - 10
		s_mon[idx] += payment

plt.title('10-12 payment')
plt.bar(['10','11','12'],s_mon, color = 'royalblue')
# plt.ticklabel_format(axis='y', style= 'plain'). plt.show()
```

```python
import matplotlib.pyplot as plt

taxi = [0,0,0]

for row in data :
	if row[-1] == '전표매입' and '택시' in row[5]:
		mon, payment = int(row[0].split('/')[1]), int(row[-3])
		idx = mon - 10
		taxi[idx] += payment

plt.title('10-12 taxi payments')
plt.plot(['10','11','12'],taxi, color = 'royalblue', label='taxi')
# plt.ticklabel_format(axis='y', style= 'plain'). plt.show()
```
