```python
import csv

with open('./data/card.csv', 'r') as f:
    reader = csv.reader(f)
    next(reader)
    data = list(reader)

print(data)
```

```python
from matplotlib import pyplot as plt

# 월별 배달음식비 지출액 그래프 그리기
deli = [0,0,0]
for row in data :
  if row[-1] == '전표매입' and row[5] == '(주)우아한형제들':
    mon, payment = int(row[0].split('/')[1]), int(row[-3])
    idx = mon-10
    deli[idx] += payment

plt.rc('font', family = 'Apple SD Gothic Neo')
plt.title('10~12 delivery payment')
plt.plot(['10','11','12'], deli, color = 'indigo', label ='delivery payment')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

taxi = [0,0,0]
deli = [0,0,0]

for row in data :
	if row[-1] == '전표매입' and '택시' in row[5]:
		mon, payment = int(row[0].split('/')[1]), int(row[-3])
		idx = mon - 10
		taxi[idx] += payment

	elif row[-1] == '전표매입' and row[5] == '(주)우아한형제들':
		mon, payment = int(row[0].split('/')[1]), int(row[-3])
		idx = mon-10
		deli[idx] += payment

plt.rc('font', family = 'Apple SD Gothic Neo')
plt.title('10-12 taxi payments')
plt.plot(['10','11','12'], taxi, color = 'royalblue', label='taxi payment')
plt.plot(['10','11','12'], deli, color = 'indigo', label ='delivery payment')
# plt.ticklabel_format(axis='y', style= 'plain').plt.show()
plt.legend()
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt
import operator
with open('./data/card.csv', 'r') as f:
    reader = csv.reader(f)
    next(reader)
    data = list(reader)

# print(data)
spending = {}
for row in data :
  if row[-1] == '전표매입':
    store, payment = row[-4], int(row[-3])
    if store not in spending.keys() :
      spending[store] = payment
    else :
      spending[store] += payment

print(spending)
top10 = sorted(spending.items(), key = operator.itemgetter(1), reverse=True)[:10]
top10_store = []
top10_amount = []
for t in top10 :
  top10_store.append(t[0])
  top10_amount.append(t[1])
plt.rc('font', family = 'Apple SD Gothic Neo')
plt.title('10-12월 지출 TOP 10')
plt.barh(top10_store, top10_amount, color = 'b')
plt.show()
```
